```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession\
        .builder\
        .master('local[*]')\
        .getOrCreate()
sc = spark.sparkContext
```

```python
lines = sc.textFile('/content/drive/My Drive/Colab Notebooks/big-data-Colab/words.txt')
```

```python
lines.count()
```

```python
words = lines.flatMap(lambda line: line.split(' '))
```

```python
tuples = words.map(lambda word : (word, 1))
```

```python
counts = tuples.reduceByKey(lambda a, b: (a + b))
```

```python
!rm -r 'SparkOutPut'
```

```python
counts.coalesce(1).saveAsTextFile('/content/drive/My Drive/Colab Notebooks/big-data-Colab/SparkOutPut')
```
